# 2D Histograms

2D histograms, also known as bivariate histograms, are used for visualizing the relationship between two numerical variables when the number of observations is big.

The plot is divided into several bins (areas) which are colored based on the number of points inside each area.

We use the **`pyplot.2dhist()`** function to create a 2D histogram.

**Example:**

```python
import numpy as np
import matplotlib.pyplot as plt

# making data
np.random.seed(1)
x = np.random.normal(size = 10000)
y = x + np.random.normal(size = 10000)

# setting figure
plt.hist2d(x, y, bins = 20)

# plotting figure
plt.show()
```

## Color palette

The default color palette is `viridis` but you can change it with the `cmap` argument, as shown in the example below.

**Example:**

```python
plt.hist2d(x, y, bins = 20, cmap = 'BuPu')

plt.show()
```

## Color transparency

You can change the transparency of the colors with `alpha` argument, which values range from 0 (transparent) to 1 (opaque, the default).

Example:

```python
plt.hist2d(x, y, bins = 20, alpha = 0.5)

plt.show()
```

## Number of bins

The number of bins or areas can be customized making use of the `bins` argument, which default to 10.

The bin specification admits four different ways listed below.

**Number of bins for the two dimensions**

If you pass an int to the argument the number will be used as the number of bins for each axis.

Example:

```python
plt.hist2d(x, y, bins = 20)

plt.show()
```

**Custom number of bins for each axis**

You can also set the number of bins for each axis with an array.

The first element will be the number of bins for the X-axis while the second will be the number of bins of the Y-axis.

Example:

```python
plt.hist2d(x, y, bins = [50, 10])

plt.show()
```

**Bin edges for the two dimensions**

An alternative is to set the bin edges for the two dimensions passing an array of edges.

Example:

```python
plt.hist2d(x, y, bins = np.arange(-3, 3, 0.2))

plt.show()
```

**Bin edges for each dimension**

The last alternative is to set the bin edges for each dimension.



```python
plt.hist2d(x, y, bins = (np.arange(-3, 3, 0.2), np.arange(-3, 3, 0.2)))

plt.show()
```
